Web Scrapper for Indeed

```python
import os
import pandas as pd
import numpy as np
import glob
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import time
import datetime as dt
```

```python

def webscrape_indeed(word1, word2, word3):

    # need to download and change the path to the one on your computer
    driverpath = "chromedriver88.exe"
    # https://sites.google.com/a/chromium.org/chromedriver/


    browserProfile = webdriver.ChromeOptions()
    browser = webdriver.Chrome(driverpath, options=browserProfile)

    browser.get('https://www.indeed.com/jobs?q={}%20{}%20{}&fromage=1&vjk='.format(word1, word2, word3))
    job_title_list = []
    address_company_list = []
    company_name_list = []
    summary_text =[]
    for row in browser.find_elements_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "jobtitle", " " ))]'):
        try:
            job_title_value = row.text
            job_title_list.append(job_title_value)
        except:
            job_title_list.append('')

    for row in browser.find_elements_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "company", " " ))]'):
        try:
            company_name_value = row.text
            company_name_list.append(company_name_value)
        except:
            company_name_list.append('')

    # find address of company
    for row in browser.find_elements_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "accessible-contrast-color-location", " " ))]'):
        try:
            address_value = row.text
            address_company_list.append(address_value)
        except:
            address_company_list.append('')

    for row in browser.find_elements_by_xpath('//*[contains(concat( " ", @class, " " ), concat( " ", "summary", " " ))]'):
        try:
            summary_value = row.text
            summary_text.append(summary_value)
        except:
            summary_text.append('')


    df = pd.DataFrame(
        {'Company': company_name_list,
         'Title': job_title_list,
         'Address': address_company_list, 
         'Summary': summary_text,
        })
    today = dt.datetime.today().strftime('%m-%d-%Y')  

    df['Date'] = today

    browser.close()
    
    return df 

# different things to run through web scrape, returns dataframe
bussiness_analyst = webscrape_indeed(word1 = 'business', word2 = 'analyst', word3='python')
```
